# Importing the libraies that to be use

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python


epl_df = pd.read_csv('/content/EPL_20_21.csv')
epl_df.head()
```

# Calculating 'MinsPerMatch' and 'GoalsPerMatch'

```python
epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype (int)

epl_df['GoalsPerMatch'] = (epl_df['Goals']/ epl_df['Matches']).astype(float)

epl_df.head()
```

# Calcuate total Goals score

```python
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

#Penalty Goals score out of the total Goals.

```python
Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

# Penalty Attempts

```python
Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

#Pie chart for penalties missed vs scored

```python
plt.figure(figsize=(8, 4))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed','Penalties scored']
colors = sns.color_palette('Set1')
plt.figure(figsize=(8, 4))
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

#Getting the average age of players in each Club

```python
# Set the figure size
plt.figure(figsize=(9,5))

# Creating a boxplot to show the distribution of players' ages by club
sns.boxplot(x='Club', y='Age', data=epl_df)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding labels and title for clarity
plt.xlabel('Club')
plt.ylabel('Age')
plt.title('Distribution of Players\' Ages by Club')

# Display the plot
plt.show()
```
